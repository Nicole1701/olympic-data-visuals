```python
import pandas as pd
import os
```

```python
athlete_data = pd.read_csv('assets/data/athlete_events.csv')
athlete_data.head()
```

```python
# selecting only columns of interest: 
athlete_data.drop(['Height', 'Weight', 'Team','Games'], axis=1, inplace=True)
athlete_data.set_index('ID')
```

```python
athlete_data.dtypes
```

```python
athlete_data.to_csv('assets/data/athlete_data_clean.csv',encoding='utf-8',index=False)
```

```python
noc_data = pd.read_csv('assets/data/noc_regions.csv')
noc_data.head()
```

```python
noc_data.rename(columns = {'region': 'country'}, inplace=True)
noc_data.head()
```

```python
# selecting only columns of interest: 
noc_data.drop(['notes'], axis=1, inplace=True)
noc_data.set_index('NOC')
```

```python
noc_data.dtypes
```

```python
noc_data.to_csv('assets/data/noc_data_clean.csv',encoding='utf-8',index=False)
```

```python
country_data = pd.read_csv('assets/data/world_country_and_usa_states_latitude_and_longitude_values.csv')
country_data.head()
```

```python
# selecting only columns of interest: 
country_data.drop(['usa_state_code', 'usa_state_latitude', 'usa_state_longitude', 'usa_state'], axis=1, inplace=True)
country_data.set_index('country_code')
```

```python
country_data.dtypes
```

```python
country_data.to_csv('assets/data/country_data_clean.csv',encoding='utf-8',index=False)
```

```python
merged_country_data = pd.merge(noc_data, country_data, on = 'country', how='left')
merged_country_data
```

```python
merged_country_data.to_csv('assets/data/merged_country_data.csv',encoding='utf-8',index=False)
```
